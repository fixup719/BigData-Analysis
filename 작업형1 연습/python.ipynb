```python
# 리스트에 있는 값 반복문을 통해 +1 한 값, 출력
# listbox = [2,4,6,8,10]
# for item in listbox:
#     # print(item + 1)
#     result = item + 1
#     print(result)


# 리스트에 있는 값 반복문을 통해 +1한 값 출력
listbox = [1,3,5,7,9]
for item in listbox:
    print(item + 1)
```

```python
# 리스트에 있는 문자열 출력
listbox = ['네모','세모','동그라미']
for i in listbox:
    print(i)
```

```python
# 리스트에 있는 문자열과 인덱스값 출력
# listbox = ['네모','세모','동그라미', '별']
# for index, item in enumerate(listbox):
#     print(index)
#     print(item)

# 리스트에 있는 문자열과 인덱스값 출력
listbox = ['가', '나', '다', '라', '마']
for idx, item in enumerate(listbox):
    print(idx, item)
```

```python
# Tip (print(f스트링))
# for index, item in enumerate(listbox):
#     print(f'인덱스{index}인 값은{item}입니다.')

for idx, item in enumerate(listbox):
    print(f'인덱스 {idx}인 값은 {item}입니다.')
```

```python
# 리턴값이 2개 (최소, 최대값을 구하는 함수)
# listbox = [15, 46, 78, 24, 56]
# def min_max(data):
#     mi = min(data)
#     ma = max(data)
#     return mi, ma

# a, b = min_max(listbox)
# print(a,b)

# 리턴값이 2개(최소, 최대 구하기)
listbox = [10,20,40,50,100]
def min_max(data):
    return min(listbox), max(listbox)

a, b = min_max(listbox)
print(a, b)
```

```python
# 평균을 구하는 함수
# listbox = [15, 46, 78, 24, 56]

# def mean(data):
#     return sum(data) / len(data)

# mean(listbox)

# 평균 구하기

list = [1,2,3,4,5]
def mean(list):
    return sum(list) / len(list)

print(mean(list))
```

### Quiz

```python
# 주어진 리스트의 최대값 - 최소값을 구하는 함수를 만드시오
# listbox = [15, 46, 78, 24, 56]

# def maxmin(data):
#     return max(data) - min(data)

# maxmin(listbox)

# 주어진 리스트의 최대-최소 값 구하는 함수
list = [10, 20, 30, 40, 50]
def maxmin(data):
    return max(list) - min(list)

maxmin(list)
```
