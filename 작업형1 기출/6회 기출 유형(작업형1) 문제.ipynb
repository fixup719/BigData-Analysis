1. 주어진 데이터는 각 소방서의 출동/도착 시간데이터이다.
출동시간과 도착시간 차이가 평균적으로 가장 오래 걸린 소방서의 시간을 분으로 변환해 출력하시오. (반올림 후 정수 출력)

```python
import pandas as pd
df = pd.read_csv("data6-1-1.csv")
```

```python
df['출동시간'] = pd.to_datetime(df['출동시간'], format = "%Y-%m-%d %H:%M:%S")
df['도착시간'] = pd.to_datetime(df['도착시간'], format = "%Y-%m-%d %H:%M:%S")
df['diff'] = abs((df['도착시간'] - df['출동시간'])).dt.total_seconds() / 60
print(round(df.groupby('소방서')['diff'].mean().sort_values(ascending=False)[0]))
```

2. 학교에서 교사 한 명당 맡은 학생 수가 가장 많은 학교를 찾고, 그 학교의 전체 교사 수를 구하시오. (정수 출력)

```python
import pandas as pd
df = pd.read_csv("data6-1-2.csv")
```

```python
# print(df.head())
df['sum'] = df.loc[:, '1학년' : '6학년'].sum(axis=1)
df['teach'] = df['sum'] / df['교사수']
df = df.sort_values('teach', ascending = False)
print(int(df.iloc[:1]['교사수']))
```

3. 연도별로 총 범죄 건수(범죄유형의 총합)의 월평균 값을 구한 후 그 값이 가장 큰 연도를 찾아, 해당 연도의 총 범죄 건수의 월평균 값을 출력하시오. (반올림하여 정수로 출력)

```python
import pandas as pd
df = pd.read_csv("data6-1-3.csv")
df['날짜'] = pd.to_datetime(df['날짜'], format = "%Y년 %m월")
df['연도'] = df['날짜'].dt.year
df['총범죄'] = df.iloc[:, 1:7].sum(axis = 1)
print(round((df.groupby('연도')['총범죄'].sum() / 12).max()))
```
