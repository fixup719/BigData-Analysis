## 문제5
- 데이터에서 IQR을 활용해 views 컬럼의 이상치를 찾고, 이상치 데이터의 수를 구하시오

```python
# your code
# IQR = 3사분위q3 - 1사분위q1

df = pd.read_csv("members.csv")

# 3사분위값
q3 = df['views'].quantile(0.75)
# 1사분위값
q1 = df['views'].quantile(0.25)
IQR = q3-q1
print(IQR)

#이상치
lower = q1 - 1.5*IQR
upper = q3 + 1.5*IQR

# 데이터수
cnt1 = sum(df['views'] < lower)
cnt2 = sum(df['views'] > upper)

print(cnt1 + cnt2)
```

## 문제6
- 'age' 컬럼의 이상치(소수점 나이와 음수나이, 0포함)를 제거하고
- 제거 전 후의 views 컬럼 표준편차를 더하시오 (최종 결과 값은 소수 둘째자리까지 출력, 셋째자리에서 반올림)

```python
# your code
df = pd.read_csv("members.csv")

# 제거전 표준편차
r1 = df['views'].std()

# 0이하 나이 제거
cond = df['age'] <= 0
df = df[~cond]

# 소수점 나이 제거 -> # 소수점 판단 ? 반올림한 값과 age가 다르면 소수점이라 판단
cond = df['age'] != round(df['age'])
df = df[~cond]


r2 = df['views'].std()

# 소수 둘째자리 까지 + 반올림
print(round(r1 + r2, 2))
```

## 문제5
- 데이터에서 IQR을 활용해 views 컬럼의 이상치를 찾고, 이상치 데이터의 수를 구하시오


```python
# 풀이
import pandas as pd
df = pd.read_csv("members.csv")

q3 = df['views'].quantile(.75)
q1 = df['views'].quantile(.25)

IQR = q3 - q1
# print(IQR)

line1 = q1 - 1.5 * IQR
line2 = q3 + 1.5 * IQR


print(sum(df['views'] < line1) + sum(df['views'] > line2))
```

```python
# 풀이
import pandas as pd
df = pd.read_csv("members.csv")

r1 = df['views'].std()

cond = df['age'] <= 0
# print(df.shape)
df = df[~cond]
# print(df.shape)

# print(df.shape)
cond = df['age'] == round(df['age'],0)
df = df[cond]
# print(df.shape)

r2 = df['views'].std()

print(round(r1 + r2, 2))
```
